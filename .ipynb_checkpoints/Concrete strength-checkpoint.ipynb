```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def create_plot(rows=len(df.columns) // 3, columns=3, columns_name=df.columns.values, kind='dist'):
    fig, ax = plt.subplots(nrows=rows, ncols=columns, figsize=(20,10))
    #print((ax.shape))
    #columns = df.columns.values
    plot = sns.distplot if kind=='dist' else sns.boxplot
    index = 0
    for i in range(3):
        plot(df[columns_name[index]], ax=ax[i,0])
        plot(df[columns_name[index+1]], ax=ax[i,1])
        plot(df[columns_name[index+2]], ax=ax[i][2])
        index += 3
    
    plt.show()
    
```
